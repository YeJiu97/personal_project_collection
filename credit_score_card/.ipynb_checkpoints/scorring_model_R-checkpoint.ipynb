```r
# 加载必要的库
library(ggcorrplot)

# 读取数据
cleaned_data <- read.csv("credit_data_clean.csv")

# 选择相关性分析的变量
continuous_vars <- cleaned_data[, c("LIMIT_BAL", "AGE", "BILL_AMT1", "BILL_AMT2", "BILL_AMT3", 
                                    "BILL_AMT4", "BILL_AMT5", "BILL_AMT6", "PAY_AMT1", 
                                    "PAY_AMT2", "PAY_AMT3", "PAY_AMT4", "PAY_AMT5", "PAY_AMT6")]

# 计算相关矩阵
cor_matrix <- cor(continuous_vars, use = "complete.obs")

# 找出高度相关的变量对（绝对值大于0.7）
highly_correlated <- which(abs(cor_matrix) > 0.7 & abs(cor_matrix) < 1, arr.ind = TRUE)

# 生成一个变量名称的向量
var_names <- colnames(cor_matrix)

# 初始化一个向量来存储要删除的变量
vars_to_remove <- c()

# 遍历高度相关的变量对，确定要删除的变量
for (i in seq_len(nrow(highly_correlated))) {
  var1 <- var_names[highly_correlated[i, 1]]
  var2 <- var_names[highly_correlated[i, 2]]
  
  # 如果var1或var2已经在删除列表中，则跳过
  if (var1 %in% vars_to_remove || var2 %in% vars_to_remove) next
  
  # 统计与其他变量高度相关的次数
  count1 <- sum(abs(cor_matrix[var1, ]) > 0.7) - 1
  count2 <- sum(abs(cor_matrix[var2, ]) > 0.7) - 1
  
  # 如果var1的高度相关次数多于var2，保留var1，删除var2；否则相反
  if (count1 >= count2) {
    vars_to_remove <- c(vars_to_remove, var2)
  } else {
    vars_to_remove <- c(vars_to_remove, var1)
  }
}

# 输出要删除的变量
print(vars_to_remove)

# 删除高度相关的变量
cleaned_data_reduced <- cleaned_data[, !colnames(cleaned_data) %in% vars_to_remove]

# 查看剩余变量
print(colnames(cleaned_data_reduced))
```

```r
# 安装必要的包
install.packages("vcd")
install.packages("corrplot")

# 加载必要的库
library(vcd)
library(corrplot)

# 选择离散变量（之前的数据集）
discrete_vars <- cleaned_data_reduced[, c("SEX", "EDUCATION", "MARRIAGE", "PAY_0", "PAY_2", "PAY_3", 
                                          "PAY_4", "PAY_5", "PAY_6")]

# 计算Cramér's V的函数
cramers_v <- function(x, y) {
  tbl <- table(x, y)
  chi2 <- chisq.test(tbl, correct = FALSE)$statistic
  n <- sum(tbl)
  min_dim <- min(dim(tbl)) - 1
  return(as.numeric(sqrt(chi2 / (n * min_dim))))
}

# 计算离散变量之间的Cramér's V相关矩阵
n <- ncol(discrete_vars)
cramer_matrix <- matrix(0, n, n)
rownames(cramer_matrix) <- colnames(discrete_vars)
colnames(cramer_matrix) <- colnames(discrete_vars)

for (i in 1:(n-1)) {
  for (j in (i+1):n) {
    cramer_matrix[i, j] <- cramers_v(discrete_vars[, i], discrete_vars[, j])
    cramer_matrix[j, i] <- cramer_matrix[i, j]
  }
}

# 打印相关矩阵
print(cramer_matrix)

# 可视化相关矩阵
corrplot(cramer_matrix, method = "circle", type = "lower", tl.col = "black", tl.srt = 45)
```

```r
# 保留 IV 值高于 0.1 的变量
selected_vars <- c("PAY_0", "PAY_4", "PAY_AMT1", "PAY_AMT6", "PAY_AMT5", "PAY_AMT4", 
                   "PAY_AMT3", "PAY_AMT2", "PAY_6", "BILL_AMT6", "LIMIT_BAL")

# 从数据集中选择这些变量
cleaned_data_final <- cleaned_data_reduced[, c(selected_vars, "default.payment")]

# 查看最终保留的变量
print(colnames(cleaned_data_final))
```

```r
# 加载必要的库
library(caret)
library(pROC)

# 确保 default.payment 被转换为因子，并重命名因子水平
cleaned_data_final$default.payment <- factor(cleaned_data_final$default.payment, levels = c(0, 1), labels = c("No", "Yes"))

# 分割数据集为训练集和测试集
set.seed(123)  # 设置随机种子以保证结果可重复
trainIndex <- createDataPartition(cleaned_data_final$default.payment, p = 0.7, list = FALSE)
train_data <- cleaned_data_final[trainIndex, ]
test_data <- cleaned_data_final[-trainIndex, ]

# 训练逻辑回归模型
model <- train(default.payment ~ ., data = train_data, method = "glm", family = binomial,
               trControl = trainControl(method = "cv", number = 10, classProbs = TRUE,
                                        summaryFunction = twoClassSummary),
               metric = "ROC")

# 打印模型结果
print(model)

# 预测测试集
predictions <- predict(model, newdata = test_data, type = "prob")
predicted_classes <- predict(model, newdata = test_data)

# 计算AUC
roc_curve <- roc(test_data$default.payment, predictions[, 2])
auc_value <- auc(roc_curve)
print(paste("AUC:", auc_value))

# 绘制ROC曲线
plot(roc_curve, main = paste("ROC Curve (AUC =", round(auc_value, 2), ")"))

# 打印混淆矩阵
conf_matrix <- confusionMatrix(predicted_classes, test_data$default.payment)
print(conf_matrix)

# 可视化混淆矩阵
fourfoldplot(conf_matrix$table, color = c("#CC6666", "#99CC99"), conf.level = 0, margin = 1, main = "Confusion Matrix")
```
