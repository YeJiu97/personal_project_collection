# 介绍

几乎所有的网站都有用户推荐计划：你可以邀请新用户尝试某个产品。通常，在新用户完成交易后，你将获得一定金额的奖励，可用作网站上的信用额度。这个挑战的目标是分析推荐计划的数据，并得出其有效性的结论。

XYZ公司于10月31日启动了一个新的推荐计划。每位推荐新用户的用户，一旦新用户购买了某样商品，就会获得10美元的信用额度。该计划已经运行了将近一个月，增长产品经理想要知道它是否成功。她非常兴奋，因为自从推荐计划开始以来，公司用户数量有了显著增长，她希望你能提供一些数据，以便她向老板展示。你能估计这个计划对网站的影响吗？

根据数据，你建议作为下一步采取什么措施？推荐计划并没有经过严格的测试，而是在指定日子对所有用户同时启动，你通过查看测试开始前后的数据来得出结论。这种方法有什么风险？你能想到更好的方式来测试推荐计划并衡量其影响吗？

**重新组织和建议**

1. **评估推荐计划的影响**：首先，通过比较推荐计划实施前后的用户增长数据、交易数量和交易总额等关键指标，评估推荐计划的直接影响。
2. **建议的下一步措施**：基于当前数据分析结果，建议进一步优化推荐计划：
   - 考虑对目标用户群体进行细分，针对不同群体设计差异化的激励措施。
   - 增强用户参与度，例如通过增加推荐奖励或提供额外奖励给表现优异的推荐用户。
3. **改进测试方法**：
   - 实施分组测试：随机将用户分为控制组和实验组，只让实验组用户知晓并参与推荐计划。
   - 设定明确的评估指标：如新用户留存率、活跃度、平均交易额等。
   - 定期评估并根据数据反馈调整推荐策略，以实现更精准的市场定位和用户增长。

通过这样系统的方法，不仅可以更准确地评估推荐计划的效果，还能根据实际效果调整和优化策略，从而提高用户参与度和企业收益。

# 导入数据集

导入常见的库：

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
warnings.filterwarnings('ignore')

%matplotlib inline
```

导入数据集：

```python
data = pd.read_csv("./../dataset/11. referral.csv", parse_dates=['date'])
data.head()
```

# 数据处理

```python
# extract date information
data['dayofweek'] = data['date'].apply(lambda x: x.dayofweek)
start_date = pd.to_datetime('2015-10-31')
```

# 数据可视化

每日金钱花费可视化：

```python
# daily money spent visualization
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(18, 12))
sns.countplot(x='date', data=data, ax=ax[0])
ax[0].set_ylabel('Daily Transaction Number', fontsize=12)

sns.barplot(x='date', y='money_spent', data=data, ax=ax[1])
ax[1].set_ylabel('Daily Transaction Amount', fontsize=12)
plt.tight_layout()
plt.show()
```
